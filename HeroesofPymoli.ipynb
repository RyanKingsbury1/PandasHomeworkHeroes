```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
heroesfile = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(heroesfile)

df.head()

```

## Player Count

* Display the total number of players


```python
# Find total number of players

number_of_players = df["SN"].nunique()

# Display total number of players

number_of_players

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of unique items

unique_items = df["Item ID"].nunique()

# Average Price

average_price = df["Price"].mean()

# Total Purchases

number_of_purchases = df["Purchase ID"].count()

# Total Revenue

total_revenue = df["Price"].sum()

# Analysis Table

analysis_df = pd.DataFrame({"Number of Unique Items" : [unique_items] , "Average Price" : [average_price]
                            , "Number of Purchases" : [number_of_purchases]
                            , "Total Revenue" : [total_revenue]})

# Display Table

analysis_df




```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Clean Gender columnn
SN_gender = df [["SN", "Gender"]]
clean_SN_gender = SN_gender.drop_duplicates("SN")
player_count = clean_SN_gender["Gender"].value_counts()

# Count genders
player_percentage = player_count/number_of_players*100
gender_breakdown = pd.DataFrame({"Percentage of Players":player_percentage
                                , "Total":player_count})
gender_breakdown["Percentage of Players"] = gender_breakdown["Percentage of Players"].map("{:.2f}".format)

# Display
gender_breakdown
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Frame
gender_purchasing = df [["Gender", "Price"]]
gender_group = gender_purchasing.groupby("Gender").sum()

# Run needed info

gender_group["Purchase Count"] = gender_purchasing.groupby("Gender").count()
gender_group["Avg Purchase Price"] = (gender_group["Price"]/gender_group
                                     ["Purchase Count"])
gender_group["Avg Purchase Total PP"] = gender_group["Avg Purchase Price"]
gender_group["Total Purchase Value"] = gender_group["Price"]

# Display
gender_group
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Make bins
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34"
             , "35-39", "40+"]

# Clean Age
SN_Age = df[["SN","Age"]]
clean_SN_Age = SN_Age.drop_duplicates("SN")

# Put in to bins

age_breakdown = pd.DataFrame({"Age":clean_SN_Age["Age"]})
age_breakdown[" "] = pd.cut(age_breakdown["Age"], age_bins, labels=bin_names)
age_group = age_breakdown.groupby(" ").sum()
age_group["Total Count"] = age_breakdown.groupby(" ").count()
age_group["Percentage of Players"] = age_group["Total Count"]/number_of_players*100

# Make and Display table

age_table = age_group[["Percentage of Players", "Total Count"]]

age_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Frame
top_spenders = df[["SN", "Price"]]

# Make Table
top_spenders_sample = top_spenders.groupby("SN").sum()
top_spenders_sample["Purchase Count"] = top_spenders.groupby("SN").count()
top_spenders_sample["Average Purchase Price"] = (top_spenders_sample
                                                ["Price"])/(top_spenders_sample["Purchase Count"])
top_spenders_sample["Total Purchase Value"] = top_spenders_sample ["Price"].map("${:.2f}".format)

#Display
top_spenders_sample.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Frame
pop_items = df [["Item ID", "Item Name", "Price"]]

# Table for Pop items
pop_items_sample = pop_items.groupby(["Item ID", "Item Name"]).sum()
pop_items_sample["Purchase Count"] = pop_items.groupby(["Item ID"
                                                       , "Item Name"]).count()
pop_items_sample["Item Price"] = (pop_items_sample["Price"]/pop_items_sample["Purchase Count"])
pop_items_sample["Total Purchase Value"] = pop_items_sample["Price"]

pop_items_table = pop_items_sample.sort_values("Purchase Count", ascending=False)

#Display
pop_items_table.head()
```
